```python
#count  of sales by category
x=df.groupby('Category')['Sales'].count()
x
```

```python
#sum  of sales by category
x=df.groupby('Category')['Sales'].sum()
x
```

```python
#state wise sales
import matplotlib.pyplot as plt
top_states = df.groupby('State')['Sales'].sum()

top_states.plot(kind='bar')

plt.title("State Wise Sales")
plt.xlabel("State")
plt.ylabel("Sales")

plt.show()
```

```python
#region was profit
region_profit = df.groupby('Region')['Profit'].sum()
print(region_profit)
```

```python
#shipping mode
df['Ship Mode'].value_counts().plot(kind='bar')
plt.title("Ship Mode Count")
plt.xlabel("Ship Mode")
plt.ylabel("Number of Orders")
plt.show()
```

```python
sub_profit = df.groupby('Sub-Category')['Profit'].sum()

print(sub_profit)
```
